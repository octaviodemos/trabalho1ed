# Modelagem do Cenário, ER e DDL (Statcast)

Notebook de apoio para documentar o cenário da tabela, a fonte de dados, o modelo entidade-relacionamento e exemplos de DDL usados nos experimentos com Delta Lake e Apache Iceberg.

## Fonte de dados utilizada

- Arquivo: `data/raw/statcast_data.csv`
- Domínio: estatísticas agregadas de arremessadores MLB Statcast
- Processamento: `src/ingestao.py` com `obter_esquema_statcast()` e `ler_e_limpar_dados()`
- Colunas finais: `nome_jogador`, `total_arremessos`, `velocidade_media`, `taxa_giro`, `media_rebatidas_contra`, `strikeouts`, `home_runs_cedidos`, `walks_cedidos`

## Modelo ER (representação)

```mermaid
erDiagram
    ARREMESSADOR {
        string nome_jogador PK
        int total_arremessos
        float velocidade_media
        int taxa_giro
        float media_rebatidas_contra
        int strikeouts
        int home_runs_cedidos
        int walks_cedidos
    }
```

A modelagem foi simplificada para uma entidade central porque o dataset é agregado por jogador na temporada.

## DDL da tabela no Apache Iceberg

```sql
CREATE TABLE ice.baseball.statcast_arremessadores (
  nome_jogador STRING,
  total_arremessos INT,
  velocidade_media FLOAT,
  taxa_giro INT,
  media_rebatidas_contra FLOAT,
  strikeouts INT,
  home_runs_cedidos INT,
  walks_cedidos INT
) USING iceberg;
```

## DDL da tabela no Delta Lake

```sql
CREATE TABLE delta.`data/delta_statcast` (
  nome_jogador STRING,
  total_arremessos INT,
  velocidade_media FLOAT,
  taxa_giro INT,
  media_rebatidas_contra FLOAT,
  strikeouts INT,
  home_runs_cedidos INT,
  walks_cedidos INT
) USING delta;
```

No experimento do notebook Delta, a criação inicial ocorre por escrita DataFrame em modo `overwrite` no caminho `data/delta_statcast`.

## DML evidenciada no trabalho

Nos notebooks `delta-lake.ipynb` e `iceberg.ipynb` estão demonstrados com exemplos práticos:

- `INSERT`: inclusão de `Pitcher, Teste`;
- `UPDATE`: correção de `velocidade_media` para `Webb, Logan`;
- `DELETE`: remoção de `Rodón, Carlos`;
- Auditoria: histórico de versões (`history`/`snapshots`).
